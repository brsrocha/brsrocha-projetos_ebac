1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
# 1, i) Importação das bibliotecas necessárias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1, ii) Carregamento do dataset
df = pd.read_csv('previsao_de_renda.csv')
````

````python
# 1, iii) Analisar posse de imóvel e posse de veículo

# Criar a figure e os eixos
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Plotar os gráficos
sns.pointplot(x='posse_de_imovel', y='renda', data=df, ax=axes[0])
sns.pointplot(x='posse_de_veiculo', y='renda', data=df, ax=axes[1])
axes[0].set_title('Posse de Imóvel x Renda')
axes[1].set_title('Posse de Veículo x Renda')

plt.show()
````

Ambas as variáveis possuem intervalos de confiança não-sobrepostos, o que é um forte indicativo que a diferença de posse de imóvel **discrimina** renda de forma **estatisticamente significante**. 
É possível notar também que, de fato, pessoas que **possuem um imóvel ou um veículo** ("True" nos gráficos) de fato possuem rendas médias estatisticamente superiores, o que faz sentido na nossa análise.
