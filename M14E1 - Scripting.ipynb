# M14E1 - Scripting
Fazendo em células separadas para organização, mas num script verdadeiro poderíamos fazer tudo na mesma célula:

```python
# Importar as libs
import pandas as pd
import matplotlib.pyplot as plt
import os

# Função fornecida na aula
def plota_pivot_table(df, value, index, func, ylabel, xlabel, opcao='nada'):
    if opcao == 'nada':
        pd.pivot_table(df, values=value, index=index,
                       aggfunc=func).plot(figsize=[15, 5])
    elif opcao == 'sort':
        pd.pivot_table(df, values=value, index=index,
                       aggfunc=func).sort_values(value).plot(figsize=[15, 5])
    elif opcao == 'unstack':
        pd.pivot_table(df, values=value, index=index,
                       aggfunc=func).unstack().plot(figsize=[15, 5])
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
```

```python
# Lista com os meses solicitados
meses = ['MAR', 'ABR', 'MAI', 'JUN', 'DEZ']
dfs = {}  # Dicionário para guardar os DataFrames

# Carregando os arquivos de cada mês
for mes in meses:
    caminho = f'input/SINASC_RO_2019_{mes}.csv'
    if os.path.exists(caminho):
        df = pd.read_csv(caminho)
        dfs[mes] = df
    else:
        print(f'Arquivo não encontrado: {caminho}')
```

Gerar gráficos pra cada mês (teste com gráfico de IDADEMAE)

```python
# Exemplo: usando QTD por 'IDADEMAE' (como na aula)
for mes, df in dfs.items():
    print(f'--- Gráficos para {mes} ---')
    plota_pivot_table(df, value='IDADEMAE', index='ESCMAE', func='count',
                      ylabel='Contagem', xlabel='Escolaridade da mãe', opcao='sort')
    plt.title(f'Contagem por Escolaridade da Mãe - {mes}')
    plt.show()
```
